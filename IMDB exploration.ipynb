```python
%%sql
SELECT * FROM title_episode LIMIT 10
```

```python
%%sql
SELECT * 
    FROM title_episode 
    WHERE parenttconst 
    LIKE 'tt0068098' 
    ORDER BY seasonnumber, episodenumber ASC
```

```python
%%sql
SELECT a.tconst, averagerating, numvotes, parenttconst, seasonnumber, episodenumber
    FROM title_ratings a
    JOIN (SELECT * 
        FROM title_episode 
            WHERE parenttconst 
            LIKE 'tt0068098'
        ) b
    ON a.tconst = b.tconst
    ORDER BY averagerating DESC 
```

```python
%%sql
SELECT primarytitle, startyear, runtimeminutes, averagerating, numvotes, seasonnumber, episodenumber 
    FROM title_basics d
    JOIN (SELECT a.tconst, averagerating, numvotes, parenttconst, seasonnumber, episodenumber
            FROM title_ratings a
            JOIN (SELECT * 
                FROM title_episode 
                    WHERE parenttconst 
                    LIKE 'tt0068098'
                ) b
            ON a.tconst = b.tconst 
         ) c
        ON d.tconst = c.tconst
    ORDER BY averagerating DESC 
```

```python
top_mash = %sql SELECT primarytitle, startyear, runtimeminutes, averagerating, numvotes, seasonnumber, episodenumber FROM title_basics d JOIN (SELECT a.tconst, averagerating, numvotes, parenttconst, seasonnumber, episodenumber FROM title_ratings a JOIN (SELECT * FROM title_episode WHERE parenttconst LIKE 'tt0068098' ) b ON a.tconst = b.tconst ) c ON d.tconst = c.tconst ORDER BY averagerating DESC 
top_mash.DataFrame().head()
```
